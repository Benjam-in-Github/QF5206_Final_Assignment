## Q1

Obtain the number of stocks in each of the GICS industries (the GICS industry is defined using 
the first 6 digits of the GICS code) for the S&P 500 universe at the beginning of 2018 and 2023
respectively. Calculate the total market cap of each industry. What are the top three industries

```python
GICS = pd.read_csv(r'C:\Users\aya18\Desktop\data_us\tickers.csv', names = ['stock', 'GICS'])
```
